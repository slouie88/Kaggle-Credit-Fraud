```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# load credit card fraud dataset
ROOT = "C:/Users/stefa/OneDrive/Kaggle/Credit Card Fraud"
FILE_PATH = os.path.join(ROOT, "creditcard.csv")

def load_dataset(csv_path=FILE_PATH):
    dataset = pd.read_csv(csv_path)
    return dataset

df = load_dataset()
df.head()
```

```python
# see structure of data
df.info()
```

```python
# count number of missing values per column in dataset
print("Number of null values per feature")
print(df.isnull().sum())
```

```python
# Descriptive statistics
df.describe()
```

```python
df.groupby(['Class']).mean()
```

```python
# Histograms for features Time through to V8
# First group of histograms for the features are when Class == 0 (non-fraudulent)
# Second group of histograms for the same features are when Class == 1 (fraudulent)

df[df['Class']==0].iloc[:,0:9].hist(bins=40, figsize=(20,15))
df[df['Class']==1].iloc[:,0:9].hist(bins=40, figsize=(20,15))
plt.show()
```

```python
# Histograms for features V9 through to V17
# First group of histograms for the features are when Class == 0 (non-fraudulent)
# Second group of histograms for the same features are when Class == 1 (fraudulent)

df[df['Class']==0].iloc[:,9:18].hist(bins=40, figsize=(20,15))
df[df['Class']==1].iloc[:,9:18].hist(bins=40, figsize=(20,15))
plt.show()
```

```python
# Histograms for features V18 through to V26
# First group of histograms for the features are when Class == 0 (non-fraudulent)
# Second group of histograms for the same features are when Class == 1 (fraudulent)

df[df['Class']==0].iloc[:,18:27].hist(bins=40, figsize=(20,15))
df[df['Class']==1].iloc[:,18:27].hist(bins=40, figsize=(20,15))
plt.show()
```

```python
# Histograms for features Amount, V27 and V28
# First group of histograms for the features are when Class == 0 (non-fraudulent)
# Second group of histograms for the same features are when Class == 1 (fraudulent)

df[df['Class']==0].iloc[:,27:30].hist(bins=40, figsize=(20,15))
df[df['Class']==1].iloc[:,27:30].hist(bins=40, figsize=(20,15))
plt.show()
```

```python
# Scatter plot of Amount vs Time
df.plot(kind="scatter", x="Time", y="Amount", label="Amount vs Time", figsize=(20,15),
c="Class", cmap=plt.get_cmap("jet"), colorbar=True)
plt.title("Amount vs Time")
plt.show()
```

```python
# Scatterplots of first 3 Principal Components
plt.figure(figsize=(20,20))



plt.subplot(3,1,1)
plt.scatter(df.iloc[:, 1], df.iloc[:, 2],
            c=df["Class"], edgecolor='none', alpha=0.5)
plt.xlabel('V1')
plt.ylabel('V2')
plt.colorbar();

plt.subplot(3,1,2)
plt.scatter(df.iloc[:, 2], df.iloc[:, 3],
            c=df["Class"], edgecolor='none', alpha=0.5)
plt.xlabel('V2')
plt.ylabel('V3')
plt.colorbar();

plt.subplot(3,1,3)
plt.scatter(df.iloc[:, 1], df.iloc[:, 3],
            c=df["Class"], edgecolor='none', alpha=0.5)
plt.xlabel('V1')
plt.ylabel('V3')
plt.colorbar();


plt.suptitle("Scatterplots of first 3 Principal Components")
plt.show()
```

```python
# 3D Scatterplot of the V13, V15, V18
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

v1_n = df[df["Class"]==0].V13
v2_n = df[df["Class"]==0].V15
v3_n = df[df["Class"]==0].V18

v1_f = df[df["Class"]==1].V13
v2_f = df[df["Class"]==1].V15
v3_f = df[df["Class"]==1].V18

ax.scatter(v1_n, v2_n, v3_n, c='r', marker='o')
ax.scatter(v1_f, v2_f, v3_f, c='b', marker='^')

ax.set_xlabel('V13')
ax.set_ylabel('V15')
ax.set_zlabel('V18')

# # rotate the axes and update
# for angle in range(0, 360):
#     ax.view_init(30, angle)
#     plt.draw()
#     plt.pause(.001)

plt.show()
```
